 The neccessary imports for our experiments to work , we'll use 
 - `numpy` to create discrete signals
 - `scipy` to perform mathematical operations like auto-convolution , auto-correlation
 - `matplotlib` to visualize the changes

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

Now , creating same discrete signals <br>
x  = discrete signal <br>
x_ = delayed discrete signal of x  <br>
y  = discrete signal  <br>
y_ = delayed discrete signal of y <br>

### Note : x ,y are same discrete signals

```python
x = np.linspace(0, 10, 20, endpoint=False)[:10]

x_ = np.linspace(0, 10, 20, endpoint=False)[1:11]

y = np.linspace(0, 10, 20, endpoint=False)[:10]

y_ = np.linspace(0, 10, 20, endpoint=False)[1:11]
```

Now , we'll make use of [correlate](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.correlate.html) and [convolve](https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.correlate.html)  with `center` mode to perform auto-convolution and auto-correlation.

```python

X = signal.convolve(x,x,'same')

Y = signal.convolve(y,y,'same')

X_ = signal.correlate(x,x_,'same')

Y_ = signal.correlate(y,y_,'same')
```

# Now , we'll plot the graph to show the differences

```python

fig , ((plt1,plt2),(plt3,plt4)) =plt.subplots(2,2)

plt1.plot(X,'ro-')
plt1.set_title('auto-convolution of x')
plt1.margins(0,2.5)

plt2.plot(Y,'go-')
plt2.set_title('auto-convolution of y')
plt2.margins(0,2.5)

plt3.plot(X_,'ro-')
plt3.set_title('auto-correlation of x')
plt3.margins(0,0.5)

plt4.plot(Y_,'go-')
plt4.set_title('auto-correlation of x')
plt4.margins(0,0.5)

plt.tight_layout()
plt.show()
```

# Now lets initialize x , y with different values to see how they change

```python
x = np.linspace(0, 10, 20, endpoint=False)[:10]

x_ = np.linspace(0, 10, 20, endpoint=False)[1:11]

y = np.linspace(0, 20, 20, endpoint=False)[:10]

y_ = np.linspace(0, 20, 20, endpoint=False)[1:11]
```

# Perform auto-convolution and auto-correlation

```python

X = signal.convolve(x,x,'same')

Y = signal.convolve(y,y,'same')

X_ = signal.correlate(x,x_,'same')

Y_ = signal.correlate(y,y_,'same')
```

# Perform Visualization

```python
fig , ((plt1,plt2),(plt3,plt4)) =plt.subplots(2,2)
plt1.plot(X,'ro-')
plt1.set_title('auto-convolution of x')
plt1.margins(0,2.5)

plt2.plot(Y,'go-')
plt2.set_title('auto-convolution of y')
plt2.margins(0,2.5)

plt3.plot(X_,'ro-')
plt3.set_title('auto-correlation of x')
plt3.margins(0,0.5)

plt4.plot(Y_,'go-')
plt4.set_title('auto-correlation of x')
plt4.margins(0,0.5)

plt.tight_layout()
plt.show()
```
